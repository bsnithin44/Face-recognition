```python
def take_picture():
    camera = cv2.VideoCapture(0)
    return_value, image = camera.read()
    image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
    cv2.imwrite('color_test2.jpeg', image)
    del(camera)
    return image
```
